## Reading Rasterio Merged with all years

```python
ds = xr.open_rasterio(file_path)
```

## Cleaning Rasterio

```python
# Band as years
da = xr.DataArray(np.arange(1999, 2020), dims=['band'])
sublist = np.arange(2017, 2019)
mask = np.isin(da, sublist, invert=True)
da = da[mask]
ds['band'] = da

# make zero negative values 
ds.values[ds.values<0] = -100
```

## Brief Visualization of Rasterio Data in Map

```python
ds.sel().hvplot.image(cmap='terrain', rasterize=True, data_aspect=2,frame_width=600)
```
